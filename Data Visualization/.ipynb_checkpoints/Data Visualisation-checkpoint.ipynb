```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

To create a chart on the same sheet

```python
%matplotlib inline 
```

Get an array between 0 and 10 

```python
x=np.arange(0,10)
```

```python
x
```

```python
y=x**2
```

```python
y
```

Create a red line chart 

```python
plt.plot(x,y, 'red')
```

Create a red color dash line

```python
plt.plot(x,y, 'r--')
```

Create plot with names

    Titile
    X axis limits 
    Y axis limits 
    X axis label 
    Y axis label

```python
plt.plot(x,y, 'g--')
plt.title("TITLE")
plt.xlim(0,12)
plt.ylim(0,100)
plt.xlabel("X label")
plt.ylabel("Y label")
```

It takes number sequence in order 

```python
mat = np.arange(0,200).reshape(20,10)
mat
```

```python
plt.imshow(mat)
```

It takes the random number do the same thing above 


```python
mat= np.random.randint(0,200, (20,10))
mat
```

Colors are distrubuted as value are distributed in random way

```python
plt.imshow(mat)
```

```python
plt.imshow(mat)
# use color bar
plt.colorbar()
```
